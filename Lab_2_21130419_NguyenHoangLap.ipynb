#6. Use scatter plot to represent the relationship between ***age*** and ***refletter.rating***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
# import matplot
import matplotlib.pyplot as plt

plt.scatter(pf['age'],pf['refletter.rating'],s=70,color="blue")
plt.title('Scatter plot represent the relationship between age and refletter.rating')
plt.xlabel('age')
plt.ylabel('refletter.rating')
plt.legend();
plt.show()
```

#7. Use scatter plot to represent the relationship between ***age*** and ***sciences.grade***
Rememer adding titles, xlabel, ylabel, ... to the plot

```python
#code
plt.scatter(pf['age'],pf['sciences.grade'],s=30,marker="*",color="green")
plt.title('Scatter plot represent the relationship between age and sciences.grade')
plt.xlabel('age')
plt.ylabel('sciences.grade')
plt.legend();
plt.show()
```

#8. Use histogram plot to represent the distribution of ***Age***
using bins=5

```python
#code
plt.hist(pf['age'],bins=5,color="green",alpha = 0.5)
plt.show()
```
